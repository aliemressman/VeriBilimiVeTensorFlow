```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
yasListesi = [10,20,30,40,40,50,60,70,80,80]
```

```python
kiloListesi = [20,60,50,80,85,90,85,86,78,92]
```

```python
numpyYasListesi = np.array(yasListesi)
numpyKiloListesi = np.array(kiloListesi)
```

```python
numpyKiloListesi
```

```python
numpyYasListesi
```

```python
plt.plot(numpyYasListesi,numpyKiloListesi,"r")
plt.xlabel("Yas")
plt.ylabel("Kilo")
plt.title("Kilonun Yaşa göre Değişim Grafiği")
plt.show()
```

```python
numpyDizisi1 = np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 ** 3
```

```python
numpyDizisi2
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"b*-")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(numpyDizisi1,numpyDizisi2,"r--")
plt.subplot(1,2,2) # 1 sıra olacak, 2 kolon olacak 2. yi çizdirecek.
plt.plot(numpyDizisi2,numpyDizisi1,"g--")
plt.show()
```

```python
benimFigure = plt.figure()
```

```python
figureAxes = benimFigure.add_axes([0.6,0.6,0.6,0.6])
figureAxes.plot(numpyDizisi1,numpyDizisi2,"r")
figureAxes.set_xlabel("X ekseni ")
figureAxes.set_ylabel("Y ekseni")
figureAxes.set_title("Grafik Başlığı")
plt.show()
```

```python
figure2 = plt.figure()

eksen1 = figure2.add_axes([0.1,0.1,0.9,0.9])
eksen2 = figure2.add_axes([0.25,0.5,0.3,0.3])

eksen1.plot(numpyDizisi1,numpyDizisi2,"g")
eksen1.set_xlabel("X ekseni ")
eksen1.set_ylabel("Y ekseni")
eksen1.set_title("Ana Grafik Başlığı")

eksen2.plot(numpyDizisi2,numpyDizisi1,"b")
eksen2.set_xlabel("X ekseni")
eksen2.set_ylabel("Y ekseni")
eksen2.set_title("Küçük Grafik Başlığı")
plt.show()
```

```python
(benimFigur,benimEksenler)= plt.subplots(nrows=1,ncols=2)

for eksen in benimEksenler:
    eksen.plot(numpyDizisi1,numpyDizisi2,"g")
    eksen.set_xlabel("X ekseni")
    
plt.tight_layout()
plt.show()
```

```python
yeniFigur = plt.figure(dpi = 100) # Loc -> boyutlandırma , dpi

yeniEksen = yeniFigur.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(numpyDizisi1,numpyDizisi1 ** 2,label = "numpyDizisi ** 2 ")
yeniEksen.plot(numpyDizisi1,numpyDizisi1 ** 3,label = "numpyDizisi ** 3 ")
yeniEksen.legend(loc=1)

plt.show()
```

```python
yeniFigur.savefig("BenimFigur.png",dpi = 200)
```
