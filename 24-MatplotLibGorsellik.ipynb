```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
numpyDizisi1 = np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 ** 2
```

```python
(benimFigur,benimEksen) = plt.subplots()
benimEksen.plot(numpyDizisi1,numpyDizisi2,color = "#3A95A8",alpha = 0.3) # aplha -> transprant
benimEksen.plot(numpyDizisi2,numpyDizisi1,color = "#C96F23")
plt.show()
```

```python
(yeniFigur,yeniEksen) = plt.subplots()
yeniEksen.plot(numpyDizisi1,numpyDizisi1 + 2, color = "blue",linewidth = 5.0,linestyle = "--")
yeniEksen.plot(numpyDizisi1,numpyDizisi1 + 3, color = "green",linewidth = 6.0,linestyle = "-")
yeniEksen.plot(numpyDizisi1,numpyDizisi1 + 4, color = "red",linestyle = ":",linewidth = 4.0,marker = "+",markersize=12) 

yeniEksen.plot(numpyDizisi1,numpyDizisi1 + 5, color = "red",linestyle = "--",marker = "o",markerfacecolor = "yellow")
plt.show()
```

## Scatter

```python
plt.scatter(numpyDizisi1,numpyDizisi2)
plt.show()
```
